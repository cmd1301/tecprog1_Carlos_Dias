```python
import pandas as pd
```

**Bora praticar!**
  
1) Utilizando o DataFrame importado (alunos3.csv) calcule a média das provas em uma nova coluna chamada (Media_provas)

```python
df = pd.read_csv('alunos3.csv', sep=';', decimal=',')
df
```

```python
df.dtypes
```

```python
from numpy import float64
df['Prova_4'] = df['Prova_4'].astype(float64)
```

```python
df['Media_provas'] = df[['Prova_1', 'Prova_2', 'Prova_3', 'Prova_4']].mean(axis=1)
df['Media_provas'] = df['Media_provas'].round(1)

df
```

2) Quem foram os alunos que obtiveram a maior e a menor média

```python
media_max = df['Media_provas'].max()
nome = df.loc[df['Media_provas'] == media_max, 'Nome'].values[0]

print(nome, media_max)
```

3) Agora una este dataframe com o `cadastro_alunos.xlsx`

```python
df2 = pd.read_excel('cadastro_alunos.xlsx')
df2
```

```python
df3 = pd.merge(df, df2, how='outer', on='RA')
df3
```

```python
cols = ['RA', 'Nome', 'Idade', 'Cidade', 'Sexo', 'Frequencia', 'Prova_1', 'Prova_2', 'Prova_3', 'Prova_4', 'Media_provas']

df3 = df3.reindex(columns=cols)

df3
```

4) Qual a média entre as Media_provas dentro do público feminino? e masculino?

```python
df3_group = df3.groupby('Sexo')

media_mf = df3_group['Media_provas'].mean()

media_mf
```

5) Qual a média de idade das pessoas que obtiveram Media_provas maior ou igual a 7?

```python
df3_maior_sete = df3[df['Media_provas'] >= 7.0]

media_idade = df3_maior_sete['Idade'].mean()

media_idade
```

6) Qual das cidades possui o maior média de Media_provas? E qual é este valor?

```python
df3_group_cidades = df3.groupby('Cidade')

media_cidades = df3_group_cidades['Media_provas'].mean()
maior_nota = media_cidades.max()

print(f'As médias de provas por cidade são: {media_cidades}.\n\n A maior nota é: {maior_nota:.2f}')
```

7. Leia o arquivo `aluno3.csv`, apenas as colunas RA, Prova_1, Prova_2, Prova_3, Prova_4 calcule a médias das notas da prova e salve em um arquivo excel (xlsx) as colunas RA e Média.


```python
cols2 = ['RA', 'Prova_1', 'Prova_2', 'Prova_3', 'Prova_4']
```

```python
df4 = pd.read_csv('alunos3.csv', sep=';', decimal=',', usecols=cols2)

df4
```

```python
df4.dtypes
```

```python
df4['Prova_4'] = df4['Prova_4'].astype(float64)
```

```python
df4['Media_provas'] = df4[['Prova_1', 'Prova_2', 'Prova_3', 'Prova_4']].mean(axis=1)
df4['Media_provas'] = df4['Media_provas'].round(1)

df4
```

```python
cols3 = ['RA', 'Media_provas']

df5 = df4[cols3]

df5.to_csv(sep=';', decimal=',')
```
